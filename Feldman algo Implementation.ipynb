## Preprocessing

```python
def replace_qs(data):
    data = data.replace(' ?', np.nan)
    data = data.dropna()
    return data

def concat_columns(data, column1="race", column2="gender", new_column="race-sex"):
    """Create a new column thats a concatenation of column1 and 2"""
    temp = []
    for index, row in data.iterrows():
        temp.append(str(str(row[column1][1:]) + "-" + str(row[column2][1:]))) # [1:] to remove whitespace prefix
    data[new_column] = pd.Series(temp)


def make_binary(data, column, privilaged):
    data[column] = data.loc[:,column].apply(lambda x: 1 if str(x) == privilaged else 0)

def one_hot_encoded(data):
    data_cat = data.select_dtypes(include=['object', 'bool'])
    data_nums = data.select_dtypes(include=['int64', 'float64'])
    data_cat_oe = pd.get_dummies(data_cat)
    data_numeric = pd.concat([data_nums,data_cat_oe], axis=1)
    return data_numeric

def pre_processing(data, dataset):
    data = replace_qs(data)
    concat_columns(data, column1="race", column2="gender", new_column="race-sex")
    if dataset=='.data':
        make_binary(data,"income", ' >50K')
    else:
        #.test file
        make_binary(data,"income", ' >50K.')
    make_binary(data,"race-sex", 'White-Male')
    make_binary(data,"race", ' White')
    make_binary(data,"gender", ' Male')
    data = data.drop(columns=["education", "fnlwgt"])
    data = one_hot_encoded(data)
    return data
    
data_numeric = pre_processing(data, '.data')
test_data_numeric = pre_processing(test_data, '.test')
```

## Average, variance, sum, count for each (numeric) feature based on race

```python
def gen_summary_df(dataset, feature):
    groupby_feature = dataset.groupby([feature])
    feature_sum_df = pd.DataFrame(groupby_feature.sum())
    feature_mean_df = pd.DataFrame(groupby_feature.mean())
    feature_var_df = pd.DataFrame(groupby_feature.var())
    count_df = pd.DataFrame(dict(dataset[feature].value_counts()), 
                 index=["count"],
                 columns=list(dataset[feature].unique())).T
    feature_groups_df = pd.concat([count_df,
                                   feature_sum_df,
                                feature_mean_df,
                               feature_var_df],
                               axis=1,
                               keys={"count":count_df,
                                   "sum": feature_sum_df,
                                     "mean":feature_mean_df,
                                     "variance":feature_var_df})
    return feature_groups_df
race_groups_df = gen_summary_df(full_dataset_edit, "race")
race_groups_df
```

```python
import seaborn as sns
sns.set_style("darkgrid")
import matplotlib.pyplot as plt

def plot_income(column, attr_groups_df, feature_name):
    """Generate a plot summarising the data"""
    income_plot = sns.barplot(x = attr_groups_df.index, y = attr_groups_df[column]["income"].values)
    if column == "sum":
        income_plot.set_title(f'Number of high earners by {feature_name}')
    elif column == "mean":
        income_plot.set_title(f'Proportion of high earners by {feature_name}')
    elif column =="variance":
        income_plot.set_title(f'Variance of the proportion of high earners by {feature_name}')
    for item in income_plot.get_xticklabels():
        item.set_rotation(45)
    return income_plot

def gen_summary_plots(feature_groups_df, feature_name):
    """Generate plots summarising the data"""
    plot_income("sum", feature_groups_df, feature_name)
    plt.show()
    plot_income("mean", feature_groups_df, feature_name)
    plt.show()
    plot_income("variance", feature_groups_df, feature_name)
    plt.show()

gen_summary_plots(race_groups_df, "race")
```

```python
from sklearn import preprocessing

x_train_std = preprocessing.StandardScaler().fit_transform(x_train) 
x_test_std = preprocessing.StandardScaler().fit_transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score,mean_squared_error,mean_absolute_error
from sklearn.metrics import accuracy_score,classification_report,plot_confusion_matrix,confusion_matrix,plot_precision_recall_curve,plot_roc_curve

def get_stats(y_test, y_pred):
    """Generate a dataframe with performance metrics"""
    results_table = pd.DataFrame({
        "Accuracy": [accuracy_score(y_test, y_pred)],
        "f1 score": [f1_score(y_test, y_pred)],
        "MSE": [np.sqrt(mean_squared_error(y_test, y_pred))],
        "MAD": [np.sqrt(mean_absolute_error(y_test, y_pred))]
    })
    #Ploting the confusion matrix
    print("Confusion matrix:\n",confusion_matrix(y_test, y_pred))
    print("Classification Report:\n",classification_report(y_test, y_pred))
    return results_table
```

## Baseline

```python
def predict_majority(y_train, y_test):
    prediction = max(y_train, key=list(y_train).count)
    return [prediction for _ in range(len(y_test))]

y_baseline_pred = predict_majority(y_train, y_test)


get_stats(y_test, y_baseline_pred)
```

The baseline model predicts the majority class all the time, and has an accuracy score of 75%. We will consider an accuracy score above this value as acceptable.

```python
def train_model(model, x_train, y_train, x_test):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    return y_pred
```

```python
def gen_gridsearch_df(gs_model):
    """Generate a pandas dataframe containing the parameters
    used for training with the mean_score for the test"""
    cols = gs_model.cv_results_["params"][0].keys()
    performance_dict = {}
    for key in cols:
        performance_dict[key] = []

    for comb in gs_model.cv_results_["params"]:
        for key in cols:
            performance_dict[key].append(comb[key])
    gridsearch_df = pd.concat([
        pd.DataFrame(performance_dict),
        pd.DataFrame(gs_model.cv_results_["mean_test_score"], columns=["mean_score"])
        ], axis=1)
    return gridsearch_df
LR_gridsearch_df = gen_gridsearch_df(LR_gs)
LR_gridsearch_df
```

```python
def get_acc_min_maj(col_interest, x_test, y_test, y_pred):
    """Get the accuracy score for the minority
    and majority groups"""
    y_test_arr = y_test.values
    maj_tot_count = 0
    maj_correct_count = 0
    min_tot_count = 0
    min_correct_count = 0
    for index,val in enumerate(x_test[col_interest]):
        # if result is for majority group
        if val == 1:
            maj_tot_count+=1
            if y_test_arr[index] == y_pred[index]:
                maj_correct_count+=1
        else:
            min_tot_count+=1
            if y_test_arr[index] == y_pred[index]:
                min_correct_count+=1
    return ((min_correct_count/min_tot_count), (maj_correct_count/maj_tot_count))

get_acc_min_maj("race", x_test, y_test, y_pred)
```

```python
def get_DI_and_CV(x_test, y_hat, sensitive_column):
    """Calulcate and optute the disparate impact and CV scores"""
    temp_data = pd.concat([x_test.reset_index() ,y_hat.reset_index() ], axis=1)
    
    s_one = x_test[sensitive_column].value_counts()[1.0]
    s_zero = x_test[sensitive_column].value_counts()[0.0]
    
    # number of occurrences in where y_hat=1 and s=1
    y_hat1_s1 = temp_data.loc[temp_data[sensitive_column] == 1, :]["income"].value_counts()[1.0]
    # s!=1
    y_hat1_not_s1 = temp_data.loc[temp_data[sensitive_column] != 1, :]["income"].value_counts()[1.0]
    # Disparate impact
    DI = (y_hat1_not_s1/s_zero)/(y_hat1_s1/s_one)
    CV = 1 - ((y_hat1_s1/s_one) - (y_hat1_not_s1/s_zero))
    
    return (DI, CV)

print(get_DI_and_CV(x_test, y_hat, 'race'))
print(get_DI_and_CV(x_test, y_hat, 'gender'))
print(get_DI_and_CV(x_test, y_hat, 'race-sex'))
```

```python
def get_fairness_stats_df(x_test, y_pred, sensitive_col, predicted_col):
    """Create a dataframe with DI and CV scores"""
    y_hat = pd.DataFrame(y_pred, columns=[predicted_col])
    DI,CV = get_DI_and_CV(x_test, y_hat, sensitive_col)
    fair_stats_df = pd.DataFrame({
        "DI":[DI],
        "CV":[CV]
    })
    return fair_stats_df
```

```python
from imblearn.over_sampling import SMOTE
def balanced_test_set(feature_to_bal, pred_col, x_test, y_test):
    """Takes in the test set as input and performs SMOTE oversampling on the
    feature to balance, e.g. race, such that the number of occurences of 1
    and 0 are equal and returns the modified, balanced test set split into
    x and y."""
    # combine x and y from split test set
    full_test_dataset = pd.concat([x_test,
                                   pd.DataFrame(y_test, 
                                                columns=["income"])],
                                  axis=1)
    # split into x and y based on the sensitive feature
    temp_y = full_test_dataset[feature_to_bal] # senstive feature column
    temp_x = full_test_dataset.drop(feature_to_bal, axis=1) # dataset without the sensitive feature
    oversample = SMOTE(random_state=101)
    # oversample so the groups in y (i.e. the sensitive attribute in this case)
    # occur an equal number of times.
    SMOTE_temp_x, SMOTE_temp_y = oversample.fit_resample(temp_x, temp_y)
    SMOTE_temp_y_df = pd.DataFrame(SMOTE_temp_y, columns=[feature_to_bal])
    SMOTE_full_dataset = pd.concat([SMOTE_temp_x, SMOTE_temp_y], axis=1)
    # separate the oversampled test set into x and y
    SMOTE_x_test = SMOTE_full_dataset.drop(pred_col, axis=1)
    SMOTE_y_test = SMOTE_full_dataset[pred_col]
    return (SMOTE_x_test, SMOTE_y_test) # return oversampled test set as a tuple, split into X and y
```

## Feldman combinatorial repair algorithm

```python
# Creating a dummy dataframe to test the repair algorithm
dummy = pd.DataFrame({
    "race":["white","white","black","other", "white", "black", "white", "other"],
    "examScores":[80,85,63,74,69,52,81,86]
})
dummy
```

```python
def gen_group_dict(data, sensitive, curr_col):
    """Group all the values of each group in a dictionary
    with key=(group name) and value=(list of values)"""
    values_dict = {}
    for i,j in zip(data[sensitive], data[curr_col]):
        if i not in values_dict.keys():
            values_dict[i] = [j]
        else:
            values_dict[i].append(j)
    return values_dict

values_dict = gen_group_dict(dummy, "race", "examScores")
print(values_dict)

def get_size_smllst_group(values_dict):
    """Find the smallest sized demographic group"""
    return min([len(arr) for arr in values_dict.values()])

size_smllst_group = get_size_smllst_group(values_dict)
```

```python
def gen_sorted_list(data, curr_col):
    """Generate a list of unique column values"""
    SL = list(set(data[curr_col].values))
    SL.sort()
    return SL

SL = gen_sorted_list(dummy,"examScores")
SL
```

```python
num_quantiles = size_smllst_group
def split_arr(arr, n):
    """Split the array into n equal parts (first part will get extra
    values if the length is odd)"""
    qot, rem = divmod(len(arr), n)
    return list(arr[i*qot+min(i, rem):(i+1)*qot+min(i+1, rem)] for i in range(n))

def gen_split_dict(dictionary, n):
    """Split the values in each group list into n sub lists
    representing the number of quantiles"""
    for key, val in dictionary.items():
        dictionary[key] = split_arr(val, n)
    return dictionary

values_dict = gen_split_dict(values_dict, num_quantiles)
values_dict
```

```python
split_arr([1,2,4,6], 5)
```

```python
import math
import copy
def median(arr):
    arr = copy.deepcopy(arr)
    arr.sort()
    return arr[math.floor((len(arr)-1)/2)]
```

```python
repair_amount = 0.5
#repair_type = "geometric"
repair_type = "combinatorial"
# python round(4.5) gives 4
def correct_round(n):
    if n - math.floor(n) < 0.5:
        return math.floor(n)
    return math.ceil(n)

def repair_col(data, values_dict, SL, repair_amount, num_quantiles, sensitive, repair_type="combinatorial"):
    new_values_dict = values_dict.copy() # e.g. {'white': [[80, 85], [69, 81]], 'black': [[63], [52]], 'other': [[74], [86]]}
    for quantile in range(num_quantiles):
        median_of_medians = []
        for arr in values_dict.values():
            median_of_medians.append(median(arr[quantile])) # append median of each quantile from each group
        target = median(median_of_medians) # target value for this quantile is the median of medians of the quantile

        for key, arr in values_dict.items():
            arr_to_repair = arr[quantile]
            new_values = []
            for val in arr_to_repair:
                # combinatorial repair
                if repair_type == "combinatorial":
                    g = SL.index(val)
                    t = SL.index(target)
                    repair_index = correct_round(g+(repair_amount*(t-g))) 
                    new_values.append(SL[repair_index])
                # Geometric repair - check
                else:
                    repair_value = ((1-repair_amount)*val + (repair_amount*target))
                    new_values.append(repair_value) # round this value????
            if quantile == 0 or key not in new_values_dict.keys():
                new_values_dict[key] = new_values
            else:
                new_values_dict[key]+=(new_values)
    new_col = []
    for instance in data[sensitive]:
        new_col.append(new_values_dict[instance].pop(0))#get the first values in the list like FIFO queue
    return new_col

repair_col(dummy, values_dict, SL, 0, num_quantiles,"race", "combinatorial")
```

```python
def run_repair(df,repair_amount, sensitive_col, repair_type):
    data = df.copy(deep=True)
    for column in [x for x in list(data.columns) if x not in [sensitive_col]]:
        SL = gen_sorted_list(data,column)
        values_dict = gen_group_dict(data, sensitive_col, column)
        
        size_smllst_group = get_size_smllst_group(values_dict)
        num_quantiles = size_smllst_group
        split_values_dict = gen_split_dict(values_dict, num_quantiles)

        data[column] = repair_col(data,split_values_dict,SL, repair_amount,num_quantiles, sensitive_col,"combinatorial")
    
    data = data.drop(sensitive_col,axis=1)
    
    return data
```

```python
def run_repair(df,repair_amount, sensitive_col, repair_type):
    """Run feldman's repair algorithm"""
    data = df.copy(deep=True)
    for column in [x for x in list(data.columns) if x not in [sensitive_col]]:
        SL = gen_sorted_list(data,column)
        values_dict = gen_group_dict(data, sensitive_col, column)
        
        size_smllst_group = get_size_smllst_group(values_dict)
        num_quantiles = size_smllst_group

        split_values_dict = gen_split_dict(values_dict, num_quantiles)

        data[column] = repair_col(data,split_values_dict,SL, repair_amount,num_quantiles, sensitive_col,"combinatorial")
        
    data = data.drop(sensitive_col,axis=1)
    # drop both columns for combined race-sex column
    if sensitive_col == "race-sex":
        data = data.drop(["race", "gender"],axis=1)
    return data


def gen_repair_scores(SENSITIVE_FEATURE, REPAIR_TYPE, INTERVAL):
    """Run the repair with a range of repair values and create
    a pandas dataframe with the resulting score metrics"""
    interval = int(INTERVAL*100)
    overall_repair_stats_df = pd.DataFrame()
    for i in range(0,100+interval,interval):
        repair = i/100
        REPAIR_LEVEL = repair


        train_check = x_train.copy(deep=True)
        test_check = x_test.copy(deep=True)

        train_check = run_repair(train_check,REPAIR_LEVEL, SENSITIVE_FEATURE, REPAIR_TYPE)
        test_check = run_repair(test_check,REPAIR_LEVEL, SENSITIVE_FEATURE, REPAIR_TYPE)

        std_scale = preprocessing.StandardScaler().fit(pd.concat([train_check, test_check]))
        x_train_check_std = std_scale.transform(train_check)
        x_test_check_std = std_scale.transform(test_check)

        y_pred = train_model(best_LR, x_train_check_std, y_train, x_test_check_std)

        min_acc, maj_acc = get_acc_min_maj(SENSITIVE_FEATURE, x_test, y_test, y_pred)
        
        repair_stats_df = pd.concat([
            pd.DataFrame({
                "train_accuracy":[accuracy_score(y_train, 
                                                train_model(best_LR, x_train_check_std, y_train, x_train_check_std))]
            }), # get the training accuracies
            get_stats(y_test, y_pred),
            get_fairness_stats_df(x_test, y_pred, SENSITIVE_FEATURE, "income"),
            pd.DataFrame({
                "accuracy_of_minority":[min_acc],
                "accuracy_of_majority":[maj_acc]
            })], axis=1)
        repair_stats_df.index = [REPAIR_LEVEL]

        overall_repair_stats_df = pd.concat([overall_repair_stats_df, repair_stats_df], axis=0)

    return overall_repair_stats_df

SENSITIVE_FEATURE = "race"
REPAIR_TYPE = "combinatorial"
INTERVAL = 0.1
race_repair_stats_df = gen_repair_scores(SENSITIVE_FEATURE, REPAIR_TYPE, INTERVAL)
```

```python
def gen_graphs(sensitive_feature, col_repair_stats_df):
    """Generate the graphs of the results of using the
    range of repair values"""
    col_accuracy_DI_CV = col_repair_stats_df[["Accuracy", "DI", "CV"]]
    ax = sns.lineplot(data=col_accuracy_DI_CV)
    ax.set(xlabel='Repair value', ylabel='Score')
    plt.title(f'Graph showing how accuracy DI and CV change\n as the repair value increases for {sensitive_feature}')
    plt.show()
    plt.clf()

    ax = sns.lineplot(data=col_accuracy_DI_CV, x="DI", y="Accuracy")
    plt.title(f'Graph showing the relationship between Accuracy and DI for {sensitive_feature}')
    plt.show()
    plt.clf()

    ax = sns.lineplot(data=col_accuracy_DI_CV, x="CV", y="Accuracy")
    plt.title(f'Graph showing the relationship between Accuracy and CV for {sensitive_feature}')
    plt.show()
    plt.clf()

    race_min_maj_accuracy = col_repair_stats_df[["accuracy_of_minority","accuracy_of_majority"]]
    ax = sns.lineplot(data=race_min_maj_accuracy)
    ax.set(xlabel='Repair value', ylabel='Accuracy')
    plt.title(f'Graph showing how the accuracy of predicting the minority\n and majority groups for {sensitive_feature} changes as the repair value increases')
    plt.show()
    
gen_graphs("race", race_repair_stats_df)
```
